```python
from IPython.display import display, Markdown
import requests
import yaml

file = '/opt/bitfusion/ami_info'
with open(file, 'r') as f:
    ami_info = yaml.load(f.read())
    
url = 'https://raw.githubusercontent.com/bitfusionio/amis/master/{}/{}/README.md'.format(ami_info['git_path'], ami_info['ami_version'])  
print(url)
r = requests.get(url)
Markdown(r.text)
```
